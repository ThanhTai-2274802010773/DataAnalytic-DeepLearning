```python
#11 Kiểm tra có giá trị NaN trong DataFrame không
df.isnull().sum()
```

```python
#12 Tìm và loại bỏ giá trị bị trùng lặp trong DataFrame dựa trên cột "Name"
df.drop_duplicates(inplace = True)

df
```

```python
#13 Thay thế giá trị lớn hơn 40 bằng giá trị trung bình của cột "Age"
df['Age'] = df['Age'].apply(lambda x: df["Age"].mean() if x > 40 else x)

df
```

```python
#14 Thêm cột BMI, tính chỉ số BMI mỗi người
df["cs_BMI"] = df["Weight"]/(df["Height"])**2

print(df)
```

```python
#15 Phân loại BMI
cs_bmi = df["cs_BMI"]

def phan_loai_bmi(cs_bmi):
    if cs_bmi < 18.5:
        return 0
    elif 18.5 <= cs_bmi < 22.9:
        return 1
    elif 23 <= cs_bmi < 24.9:
        return 2
    else:
        return 3

df["PhanLoai"] = df["cs_BMI"].apply(phan_loai_bmi)
df
```

```python
#17 
x = df["Name"]
def gioi_tinh(x):
    if "Mr" in x:
        return "Male"
    elif "Mrs" in x or "Miss" in x:
        return "Female"
    else:
        return "Unknow"

df["GT"] = df["Name"].apply(gioi_tinh)

df["Name"] = df["Name"].str.replace(r"(Mr\.|Miss\.|Mrs\.)\s*", "", regex=True)
df
```

### **1.1 TẠO MẢNG VỚI NUMPY**

```python
import numpy as np
#Tạọ mảng 1 chiều:
mang_1_chieu = np.array([1,3,5,7,9])
print(mang_1_chieu)

#Tạo mảng 2 chiều:
mang_2_chieu = np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(mang_2_chieu)

#Tạo mảng ngẩu nhiên:
mang_ngau_nhien = np.random.rand(3, 3)
print(mang_ngau_nhien)
```

### Các thao tác cơ bản của numpy

```python
mang_1_chieu = np.array([1,3,5,7,9])

#Cộng trừ nhân chia với một số:
print("Cộng thêm 10:", mang_1_chieu + 10)
print("Trừ đi 10:", mang_1_chieu - 10)
print("Nhân lên 10:", mang_1_chieu * 10)
print("Chia cho 10:", mang_1_chieu / 10)

#Tính tổng, trung bình, lớn nhất, nhỏ nhất:
print("Tính tổng:", mang_1_chieu.sum())
print("Trung bình:", mang_1_chieu.mean())
print("Lớn nhất:", mang_1_chieu.max())
print("Nhỏ nhất:", mang_1_chieu.min())

#Lấy phần tử bất kỳ trong mảng:
print("Lâý số 3 trong mảng:", mang_1_chieu[1])
```

### **Pandas**

```python
import pandas as pd 
data = {
    "Tên": ["Anh", "Dat", "Tai", "Linh"],
    "Tuổi": [30, 50, 21, 100],
    "Điểm": [55, 85, 100, 95]
}
df = pd.DataFrame(data)
print(df)
```

```python
#In ra cột tuổi
print(df["Tuổi"])

#In ra hàng đầu tiên
print(df.iloc[0])

#In ra dữ liệu có điểm >80
print(df[df["Điểm"] > 80])

#Tính trung bình điểm lớp học
print(df["Điểm"].mean())
```

### **Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
#Vẽ biểu đồ đường:
x = [1,3,5,7,8]
y = [3,7,9,8,7]
plt.plot(x, y, label="Biểu đồ đường", color="blue", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.legend()
plt.show()
```

```python
#Biểu đồ cột:
Ten = ["Ly", "Mai", "Xuan", "Bich", "An", "Khang", "Huy"]
Diem = [5,7,1,5,8,9,10]
plt.bar(Ten, Diem, color="blue")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.legend()
plt.show()
```

```python
#Biểu đồ phân tán:
x = np.random.rand(100)
y = np.random.rand(100)
size = np.random.rand(100) * 100
plt.scatter(x, y, size, alpha = 0.5, c = "pink")
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```
